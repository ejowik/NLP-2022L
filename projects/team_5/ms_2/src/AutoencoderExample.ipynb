# 1. Example 1

The first example is a simple embedding of random walks in a graph. The graph is connected in such a way, that points close in the embedding space are connected together. The example aims to show, that the encoder can in fact encode structure of the graph. As we will see, the embedding will produce clusters, that correspond to the parts of the graph.

## 1.1 Data Preparation

Firstly, we generate the following graph:

```python
def generate_example_1(n=100):
    group_structure = torch.randn(n,2)
    A = torch.zeros(n,n)
    for i in range(n):
        d = [(group_structure[i] - group_structure[j]).pow(2).sum() for j in range(n)]
        A[i,sorted(range(n), key=lambda x: d[x])[1:3]]=1
    return group_structure, A
```

```python
torch.random.manual_seed(3)
group_structure, A = generate_example_1()
```

```python
plt.scatter(group_structure[:,0], group_structure[:,1])
for i in range(n):
    for j in range(n):
        if A[i,j]:
            a = group_structure[i]
            b = group_structure[j]
            plt.plot([a[0], b[0]], [a[1], b[1]], c='skyblue')
```

Then, a dataset of random walks is being created:

```python
def sample_path(A, n):
    A = A == 1
    v = np.random.randint(A.shape[0])
    p = [v]
    for j in range(n-1):
        available = set(np.arange(A.shape[1])[A[v, :]]).difference(set(p))
        if len(available) == 0:
            v = np.random.choice(np.arange(A.shape[1])[A[v, :]])
        else:
            v = np.random.choice(list(available))
        p.append(v)
    return p
```

```python
X = torch.stack([
    group_structure[sample_path(A, 16)]
    for _ in range(10000)
])
```

Each path is 16 in length. An example is shown below:

```python
plt.scatter(group_structure[:,0], group_structure[:,1])
for i in range(n):
    for j in range(n):
        if A[i,j]:
            a = group_structure[i]
            b = group_structure[j]
            plt.plot([a[0], b[0]], [a[1], b[1]], c='skyblue')
x = X[0]
plt.scatter(x[:,0], x[:,1])
```

```python
ds = torch.utils.data.TensorDataset(X)
dl = torch.utils.data.DataLoader(ds, batch_size=64, shuffle=True)
```

## 1.2 Training

```python
m = Autoencoder(2,8,10,2,3,variational=False, max_log2len=16)
Ls = []
```

```python
opt = torch.optim.Adamax(m.parameters(), lr=0.001, weight_decay=0.01)
```

```python
for _ in range(10):
    pbar = tqdm(dl)
    for (x,) in pbar:
        opt.zero_grad()
        p = m(x)
        L = (x - p).pow(2).mean()
        L.backward()
        opt.step()
        pbar.set_description(f"L: {L.item()}")
        Ls.append(L.item())
```

```python
plt.plot(np.log(Ls))
```

## 1.3 Results

Firstly, we show the embeddings using TSNE. 

```python
encoded = m.encoder(X[0:2000])
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=30).fit_transform(encoded.detach().numpy())
kmeans = KMeans(n_clusters=11).fit(encoded.detach().numpy())
```

```python
plt.scatter(X_embedded[:,0], X_embedded[:,1], c=kmeans.labels_) # ind.numpy())
plt.title("First Vertex" )
a1 = -2.2
b1 = 10
a2 = 0
b2 = -75
```

As we can see, we obtained clear clusters of sequences. Below we show a particular cluster from the representation:

```python
plt.scatter(group_structure[:,0], group_structure[:,1])
for i in range(n):
    for j in range(n):
        if A[i,j]:
            a = group_structure[i]
            b = group_structure[j]
            plt.plot([a[0], b[0]], [a[1], b[1]], c='skyblue')
for x in X[:2000][kmeans.labels_ == 5]:
    plt.scatter(x[:,0], x[:,1], c='r')
```

The obsevations in the cluster walked only through the selected nodes. As we can see, the cluster is a particular topological neighbourhood in the graph. Interestingly, the observations that are close in the original representation, but are distant in the topology of the graph are not selected. This shows, that in fact the algorithm used the positions of the nodes more as identifiers, than what they originally represented.

This task has, however, one disadvantage. The nodes in the random walk are in fact close in the original representations. What will happen, if the walks will be non-local? 

# 2. Example 2

The following example explores a model of groups of synonyms. Synonyms are clusters of data in euclidean space. The random walks never go through the same cluster twice in a row.

## 2.1 Data Preparation

```python
def generate_example_2():
    group_structure = torch.stack([
        (torch.randn(1,5).T + 0.05*torch.randn(1,5)).flatten(),
        (torch.randn(1,5).T + 0.1*torch.randn(1,5)).flatten()]
    ).T
    A = torch.ones(25,25)
    for i in range(5):
      A[i*5:(i+1)*5, i*5:(i+1)*5] = 0
    return group_structure, A
```

```python
torch.manual_seed(3)
group_structure, A = generate_example_2()
plt.scatter(group_structure[:,0], group_structure[:,1])
```

```python
X = torch.stack([group_structure[sample_path(A, 5)] for _ in range(10000)])
```

```python
ds = torch.utils.data.TensorDataset(X)
dl = torch.utils.data.DataLoader(ds, batch_size=64, shuffle=True)
```

Clusters of data are generated. The random walks jump between the groups. In this case, the walks are highly non-local, and very opposite to the previous example. 

## 2.2 Training 

```python
m = Autoencoder(2,8,10,2,3,max_log2len=8,dropout=0)
Ls = []
```

```python
opt = torch.optim.Adamax(m.parameters(), lr=0.0001, weight_decay=0.01)
```

```python
for _ in range(10):
    pbar = tqdm(dl)
    for (x,) in pbar:
        opt.zero_grad()
        p = m(x)
        L = (x - p).pow(2).mean()
        L.backward()
        opt.step()
        pbar.set_description(f"L: {L.item()}")
        Ls.append(L.item())
```

```python
plt.plot(np.log(Ls))
```

The autoencoder has been trained to a quite good accuracy. Again, firstly we analyse the results using TSNE.

```python
encoded = m.encoder(X)
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=30).fit_transform(encoded[:2000].detach().numpy())
```

```python
plt.scatter(X_embedded[:,0], X_embedded[:,1], c=kmeans.labels_) # ind.numpy())
plt.title("First Vertex" )
a1 = -2.2
b1 = 10
a2 = 0
b2 = -75
```

While clusters of data are in fact visible in the representation, they are not as distinct as in the previous example. We will now show, that the similarity between sequences is very sensible. The similarity is calculated as a cosine between vectors.

```python
_, most_similar = torch.sort((encoded @ encoded[0])/torch.sqrt((encoded * encoded).sum(1)), descending=True)
```

```python
plt.scatter(group_structure[:,0], group_structure[:,1])
for i,x in enumerate(X[most_similar[0:10]]):
  plt.plot(x[:,0], x[:,1], label=str(i))
```

This example shows an observation with its closest 10 neighbours in the representation. As we can see, all the walks form a very similar arrow-like structure. 

```python
plt.scatter(group_structure[:,0], group_structure[:,1])
for i,x in enumerate(X[most_similar[10:20]]):
  plt.plot(x[:,0], x[:,1], label=str(i))
```

The next 10 closest neighbours shown above, are different, but still hold similarity to the arrow-like shape.
